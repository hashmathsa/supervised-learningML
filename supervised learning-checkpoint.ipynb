##b.Feature Scaling

##Train_Test_Split

```python
# Importing train_test_split to split the dataset 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_scaled,df.target,test_size=0.2,random_state=0)
```

##Classification Algorithm Implementation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize classifiers
logreg = LogisticRegression(random_state=42)
dtree = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
svm = SVC(random_state=42)
knn = KNeighborsClassifier()
```

```python
# Fit models to the scaled data using LogisticRegression
logreg.fit(x_train, y_train)
```

```python
# Fit models to the scaled data using DecisionTreeClassifier
dtree.fit(x_train, y_train)
```

```python
# Fit models to the scaled data using RandomForestClassifier
rf.fit(x_train, y_train)
```

```python
# Fit models to the scaled data using SVC
svm.fit(x_train, y_train)
```

```python
# Fit models to the scaled data using KNeighborsClassifier
knn.fit(x_train, y_train)
```

```python
# Predictions on test set of LogisticRegression model
y_pred_logreg = logreg.predict(x_test)
```

```python
# Predictions on test set of DecisionTreeClassifier model
y_pred_dtree = dtree.predict(x_test)
```

```python
# Predictions on test set of RandomForestClassifier model
y_pred_rf = rf.predict(x_test)
```

```python
# Predictions on test set of SVC model
y_pred_svm = svm.predict(x_test)
```

```python
# Predictions on test set of KNeighborsClassifier model
y_pred_knn = knn.predict(x_test)
```

##An overview of each algorithm, along with a brief rationale:

A linear model that forecasts the likelihood of the target class is called a logistic regression. appropriate for this dataset because the problem is binary classification.

A tree-based model called a decision tree classifier divides the data into subgroups according to features. It is appropriate since it is interpretable and can manage non-linear relationships.

A multi-decision tree ensemble model known as the Random Forest Classifier. Appropriate because it enhances generalization and lessens overfitting.

Support Vector Machine (SVM): A model based on kernels that determines the best hyperplane for class separation. It is appropriate since it is noise-resistant and able to manage high-dimensional data.

A distance-based model called k-Nearest Neighbors (k-NN) uses the neighbors closest to it to forecast the target class. It works well because it is easy to use and can manage non-linear relationships.


##Model Comparison

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Initialize lists to store the results
results = []
# Evaluate performance
models = [logreg, dtree, rf, svm, knn]
model_names = ["Logistic Regression", "Decision Tree", "Random Forest", "SVM", "k-NN"]
for model, name in zip(models, model_names):
    acc = accuracy_score(y_test, model.predict(x_test))
    prec = precision_score(y_test, model.predict(x_test))
    rec = recall_score(y_test, model.predict(x_test))
    f1 = f1_score(y_test, model.predict(x_test))
    results.append((name, acc, prec, rec, f1))
    print(f"{name}: Accuracy={acc:.4f}, Precision={prec:.4f}, Recall={rec:.4f}, F1-score={f1:.4f}")
```

```python
from sklearn.metrics import classification_report
# Evaluate performance for each model
for model, name in zip(models, model_names):
    y_pred = model.predict(x_test)
    report = classification_report(y_test, y_pred, target_names=["Benign", "Malignant"])
    print(f"{name}:\n{report}\n")
```

```python
from sklearn.model_selection import cross_val_score
# Initialize classifiers (same as before)
models = [logreg, dtree, rf, svm, knn]
# Evaluate performance using cross-validation
for model, name in zip(models, model_names):
    scores = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    avg_accuracy = np.mean(scores)
    print(f"{name}: Average Accuracy = {avg_accuracy:.4f}")
```

```python
results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-score"])

# Find the best and worst models based on F1-score
best_model = results_df.loc[results_df["F1-score"].idxmax()]
worst_model = results_df.loc[results_df["F1-score"].idxmin()]

print(f"\nBest Model: {best_model['Model']} with F1-score={best_model['F1-score']:.4f}")
print(f"Worst Model: {worst_model['Model']} with F1-score={worst_model['F1-score']:.4f}")
```

##Even though every model did a fantastic job, Logistic Regression, Decision Tree, Random Forest, and SVM stand out as the best because they all achieved flawless ratings in every category. However, even though it is the poorest, k-NN performs well with a high F1-score, suggesting that it is a viable alternative albeit marginally less ideal than the others.
